#  Shallow Net in Keras

Building a shallow neural network to classify MNIST digits

#### Set seed for reproducibility

```python
import numpy as np
np.random.seed(42)
```

####  Load dependencies

```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
```

#### Load data

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
y_train[0:10]
```

#### Preprocess data
##### Inputs as one dimensional array of normalized floats 

```python
X_train_flat = X_train.reshape(60000,784).astype('float32')
X_test_flat = X_test.reshape(10000,784).astype('float32')
```

```python
X_train_flat /= 255
X_test_flat /= 255
```

#####  Output Hot encoding

```python
n_classes = 10
y_train_cat = keras.utils.to_categorical(y_train)
y_test_cat = keras.utils.to_categorical(y_test)
```

```python
y_train[0]
```

```python
y_train_cat[0]
```

#### Design neural network architecture

```python
model = Sequential()
model.add(Dense((64), activation='sigmoid', input_shape=(784,)))
model.add(Dense((10), activation='softmax'))
```

```python
model.summary()
```

#### Configure model 

```python
model.compile(loss='mean_squared_error', optimizer=SGD(lr=0.01),metrics=['accuracy'])
```

#### Train

```python
model.fit(X_train_flat, y_train_cat, batch_size=128, epochs=300, verbose=1, validation_data=(X_test_flat,y_test_cat))
```
